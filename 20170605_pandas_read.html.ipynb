## scraping web table using pandas.read_html

last update: Jun 5. 2017

When you want to scrape some website, you can use pandas.read_html instead of BeautifulSoup and requests

* __Import pandas__

```python
import pandas as pd
```

* __By using read.html you can scrape what you want__
* __In this case, website that you want to scrape must contain table data__
* __assign finance.naver.com to html_df__

```python
html_df = pd.read_html("http://finance.naver.com/item/board.nhn?code=036930")
```

```python
html_df
```

* __first data frame is__

```python
html_df[0]
```

* __second data frame is__

```python
html_df[1]
```

* __this data frame is what you want to clean up__
* __therefore, you can use dropna() method to remove "NaN" in data frame__

```python
html_df = html_df[1].dropna()
html_df.head(10)
```

* __as you know, the header of html_df is [0, 1, 2, 3, 4,...]__
* __by using html_df.columns = html_df.iloc[0], you can replace columns name__

```python
html_df.columns = html_df.iloc[0]
html_df.head(10)
```

* __remove first row by using [drop index or [1:]] __

```python
html_df.drop(html_df.index[0]).head(10)
```

```python
html_df[1:].head(10)
```

* __It is very short code ! __

```python
html_df = html_df[1].dropna()
html_df.columns = html_df.iloc[0]
html_df = html_df[1:]
```
