## __3. iris data를 이용한 연습__ ##
---------------------------------------

* iris data import

```python
from sklearn import datasets
iris = datasets.load_iris()                                                    ## sklearn으로부터 example dataset 불러오기
data = pd.DataFrame(iris['data'])                                     ## array를 dataframe으로 만들기
target = pd.DataFrame(iris['target'])
target_names = pd.DataFrame(iris['target_names'])

target['species'] = ""
for i in range(len(target)):                                                 
    if target[0][i] == 0:
        target['species'][i] = 'setosa'
    elif target[0][i] == 1:
        target['species'][i] = 'versicolor'
    elif target[0][i] == 2:
        target['species'][i] = 'virginica'

del target[0]

iris = pd.concat([data, target], axis=1)                             ## dataframe으로 만들기

iris = iris.rename(columns = {0:'sepal_length', 1:'sepal_width', 2:'petal_length', 3:'petal_width'})
```

```python
iris
```

* __기초통계량__ 구하기

```python
iris.describe()
```

```python
iris.sepal_length.mean()
```

```python
iris['sepal_length'].mean()
```

* 소수점 2자리만 표현

```python
print("Sepal Length: %.2f" % (iris['sepal_length'].mean()))
```

* setosa인 species만 선택하기

```python
setosa = iris[iris.species == 'setosa']
```

```python
setosa
```

* sepal length가 5.0을 넘는 setosa를 선택하기

```python
iris[(iris['species'] == 'setosa') & (iris['sepal_length'] > 5.0)]
```

* __seaborn을 이용한 다양한 plots__

```python
import seaborn as sns
%matplotlib inline
## 그래프를 셀 안에서 보여주기
```

```python
sns.jointplot(x = setosa.sepal_length, y = setosa.sepal_width, kind="reg")
```

```python
sns.jointplot(x = setosa.sepal_length, y = setosa.sepal_width, kind="hex")
```

```python
sns.jointplot(x = setosa.sepal_length, y = setosa.sepal_width, kind="kde")
```

## __5. FOR 문__ ##
--------------------
* iris에서 species가 setosa면 종에 세토사를, virginica면 버지니카를 입력

```python
iris['종'] = ""
```

```python
for i in range(len(iris)):                   ## iris데이터를 0부터 149까지 세기(파이썬은 0부터 시작)
    if iris['species'][i] == 'setosa':   ## i번째 species가 'setosa'면
        iris['종'][i] = '세토사'              ## i번째 종 행에 '세토사'를 입력
    else:                                           ## 그렇지 않으면 (즉, setosa가 아니면)
        iris['종'][i] = 'not세토사'          ## 'not 세토사'를 입력
```
